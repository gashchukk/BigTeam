```python
spark.sql(f"""USE CATALOG sandwich""")
spark.sql("""USE SCHEMA silver""")
```

```python
customer_id = 446044
query = f"""
SELECT SUM(l_quantity) AS total_items FROM lineitem
JOIN orders ON lineitem.l_orderkey = orders.o_orderkey
WHERE orders.o_custkey = {customer_id}
"""

print(f"Total items customer {customer_id} bought: {spark.sql(query).collect()[0]['total_items']}")
```

```python
customer_id = 564787

query = f"""
SELECT MIN(o_orderdate) AS first_buy_date FROM orders
WHERE o_custkey = {customer_id};
"""
spark.sql(query).collect()
print(f"First buy date for customer {customer_id}: {spark.sql(query).collect()[0]['first_buy_date']}")
```

```python
customer_id = 85909 

query = f"""
SELECT COUNT(*) AS total_orders
FROM orders
WHERE o_custkey = {customer_id}
"""
print(f"Total number of orders for customer {customer_id}: {spark.sql(query).collect()[0]['total_orders']}")
```

```python
query = f"""
SELECT nation.n_name AS nationality FROM customer
JOIN orders ON customer.c_custkey = orders.o_custkey
JOIN nation ON customer.c_nationkey = nation.n_nationkey
GROUP BY nation.n_name
ORDER BY COUNT(*) DESC
LIMIT 1
"""
print(f"Ð¡ustomer who made the most orders in total is from {spark.sql(query).collect()[0]['nationality']}")
```
