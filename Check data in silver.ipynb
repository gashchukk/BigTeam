Check completeness of the tables

```python
bronze_schema = "sandwich.bronze"
silver_schema = "sandwich.silver"

tables = [row.tableName for row in spark.sql(f"SHOW TABLES IN {bronze_schema}").collect()]

for tbl in tables:
    bronze_count = spark.sql(
        f"SELECT COUNT(*) as cnt FROM {bronze_schema}.{tbl}"
    ).collect()[0]['cnt']
    silver_count = spark.sql(
        f"SELECT COUNT(*) as cnt FROM {silver_schema}.{tbl}"
    ).collect()[0]['cnt']
    
    if bronze_count == silver_count:
        print(f"[OK] Table {tbl}: row count matches ({bronze_count})")
    else:
        print(f"[ERROR] Table {tbl}: bronze={bronze_count}, silver={silver_count}")
```
