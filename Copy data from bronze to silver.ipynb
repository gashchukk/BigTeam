```python
bronze_schema = "sandwich.bronze"
silver_schema = "sandwich.silver"

tables_names = ["customer", "orders", "lineitem", "part", "supplier", "partsupp", "nation", "region"]

def copy_data(tbl_name):
    print(f"Copying data: {tbl_name}")
    spark.sql(f"""
        INSERT INTO {silver_schema}.{tbl_name}
        SELECT * FROM {bronze_schema}.{tbl_name}
    """)

for tbl in tables_names:
    copy_data(tbl)

print("All data copied from bronze to silver successfully!")
```
